#Filtrando e mesclando bases dos resultados do último mês

Filtrando pelo último resultado

```python
ultima_mes = df1['Data_Referencia'].max()
df1 = df1.loc[df1['Data_Referencia'] == ultima_mes]
df2 = df2.loc[df2['Data_Referencia'] == ultima_mes]
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
```

Filtrando pelos tickers do FIIs. Para isso, necessito dos tickers dos FII que não encontrei ou são fornecidos pela CVM.

Fonte: https://www.infomoney.com.br/planilhas/cnpj-das-empresas-listadas-na-b3-para-declarar-acoes-no-imposto-de-renda/

```python
FII = pd.read_csv('/content/lista_de_fiis.csv',sep=';',  encoding='ISO-8859-1',index_col=False)
```

```python
FII
```

Adicionando FII que não está na lista acima (VGIA11)

```python
nova_linha = ['VALORA CRA FIAGRO','VGIA11','41.081.088/0001-09']
FII = FII.append(pd.Series(nova_linha, index=FII.columns), ignore_index=True)
```

```python
FII = FII.rename(columns={'CNPJ': 'CNPJ_Fundo'})
FII = FII[['Nome_Fundo', 'CNPJ_Fundo', 'Código']]
```

```python
df_merged = pd.merge(FII, df1, on='CNPJ_Fundo',how='outer')
df_merged = pd.merge(df_merged, df2, on='CNPJ_Fundo',how='outer')
df_merged = df_merged.dropna()
```

```python
# Adicionar o sufixo '.SA' à coluna 'Código'
df_merged['Código'] = df_merged['Código'].apply(lambda x: x + '.SA')
```

```python
df_merged = df_merged.drop(['Nome_Fundo_y','Data_Referencia_y'], axis =1 )
df_merged = df_merged.rename(columns={'Nome_Fundo_x': 'Nome_Fundo', 'Data_Referencia_x':'Data_Referencia'})
```

```python
df_merged.head(10)
```

#Filtrando os FII da minha carteira pessoal

```python
carteira_fii = ['KISU11.SA','RINV11.SA','HGLG11.SA','BTLG11.SA','VISC11.SA','MALL11.SA','JSRE11.SA','HGRE11.SA',
                'HGRU11.SA','FIIB11.SA','MCCI11.SA','RRCI11.SA','IRDM11.SA','BDIF11.SA','VGIA11.SA']
```

```python
df_filtrado = df_merged.loc[df_merged['Código'].isin(carteira_fii)]
df_filtrado = df_filtrado.reset_index(drop=True)
df_filtrado = df_filtrado[['Data_Referencia','Código','Valor_Patrimonial_Cotas','Percentual_Dividend_Yield_Mes']]
```

```python
df_filtrado['Valor_Patrimonial_Cotas'] = pd.to_numeric(df_filtrado['Valor_Patrimonial_Cotas'])
```

```python
df_filtrado
```

```python
df_filtrado.to_csv('fii.csv', index=False)
```
